```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Student_Performance.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
y = df['Performance Index']
y
```

```python
x = df.drop('Performance Index', axis=1)
x
```

```python
x['Extracurricular Activities'] = x['Extracurricular Activities'].replace('Yes', 1)
x['Extracurricular Activities'] = x['Extracurricular Activities'].replace('No', 0)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=40, test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
print('MSE: ', mean_squared_error(y_test, y_pred))
print('R2 Score: ', r2_score(y_test, y_pred))
```

```python
student1_input = [8,92,1,7,10] #	Hours Studied, Previous Scores, Extracurricular Activities, Sleep Hours, Sample Question Papers Practiced

student1_performance_index = model.predict(np.array([student1_input]))
print('Predicted performance index of Student 1: ', student1_performance_index)
```

```python
student2_input = [4,88,0,9,6]

student2_performance_index = model.predict(np.array([student2_input]))
print('Predicted performance index of Student 2: ', student2_performance_index)
```
