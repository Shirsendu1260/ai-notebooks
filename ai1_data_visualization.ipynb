```python
# As 'subsidiary_of' column has majority of null values, we will drop it.
df2_final = df2.iloc[:, 0:6]
df2_final
```

```python
# Group by'sector' and calculate the mean of 'revenue' and 'employees' for each sector
df2_grouped_sector = df2_final.groupby('sector').agg({'revenue': 'mean', 'employees': 'mean'}).reset_index()

print('Mean of Revenue and No. of Employees by Sector:\n')
df2_grouped_sector
```

```python
# Group by 'office_location'
df2_grouped_office = df2_final.groupby('office_location')
df2_grouped_office
```

```python
df2_grouped_office = df2_grouped_office.agg({'revenue': ['sum','mean','count'], 'employees': ['sum','mean','count']}).reset_index()

print('Statistics by Office Location:\n')
print(df2_grouped_office)
```

```python
df2_grouped_office.columns
```

```python
# Plot bar plot on total no. of employees based on office location
plt.figure(figsize=(18,10))
sns.barplot(data=df2_grouped_office, x='office_location', y=('employees','sum'), color='green')
plt.xlabel('Office Location')
plt.ylabel('Total no. of employees')
plt.title('Bar plot on total no. of employees based on office location')
plt.show()
```

```python
# Group by 'year_established' and count the number of companies established each year
df2_grouped_year = df2_final.groupby('year_established').size().reset_index(name='count')

print('Total number of companies established each year:\n')
print(df2_grouped_year)
```
