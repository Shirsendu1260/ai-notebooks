```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/Asian_Medal tally.csv')
df2 = pd.read_csv('/content/pokemon.csv')
```

Analyzing **Asian_Medal tally.csv** dataset

```python
df1.info()
df1.describe()
```

```python
df1.isnull().sum()
```

```python
df1_new = df1.head(10)
```

```python
# draw pie plot to visualize 2018 Summer Olympic achievements by countries
plt.figure(figsize=(14, 10))
country_data = df1_new['NOC']
medal_data = df1_new['Total']
plt.pie(medal_data, labels=country_data, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Gold medal achievements of ten most successful\n" + "countries in 2018 Summer Olympics")
plt.show()
```

```python
# Draw pie plot based on 'attack' and 'defense' using Seaborn
g1 = df2_new.loc[df2_new.generation==1,:]
sns.scatterplot(x='attack', y='defense', data=g1)
```

```python
g1 = df2_new.loc[df2_new.generation==1,:]
g1.plot.scatter('attack', 'defense')
```

```python
# Draw pie plot based on 'attack' and 'defense' using Matplotlib
plt.scatter('attack', 'defense', data=g1)
```

```python
# Plot bubble plot on our derived 'g1' dataset
plt.scatter(data=g1, x='attack', y='defense', s=g1['weight_kg']*10, alpha=0.4)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Pokemon Attack vs Defense')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df2_new['attack'], y=df2_new['defense'], s=df2_new['weight_kg']*10, alpha=0.4)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Pokemon Attack vs Defense')
plt.grid(True)
plt.show()
```

```python
# Draw scatter plot to show legendary=1 and legendary=0 point distribution based 'attack' and 'defense'
plt.figure(figsize=(10, 6))
plt.scatter('attack', 'defense', data=df2_new[df2_new.is_legendary==1], marker='o', color='blue')
plt.scatter('attack', 'defense', data=df2_new[df2_new.is_legendary==0], marker='o', color='orange')
plt.legend(('Yes', 'No'), title='Is legendary?')
plt.show()
```

```python
# Draw histogram to show frequency distributions of 'attack'
attack_data = df2_new['attack']

plt.hist(attack_data, bins=40, color='lightgreen', edgecolor='black')
plt.xlabel('Attack')
plt.ylabel('Frequency')
plt.title('Histogram of Attack Attribute')
plt.grid(True)
plt.show()
```

```python
plt.hist(attack_data, bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Attack')
plt.ylabel('Frequency')
plt.title('Histogram of Attack Attribute')
plt.grid(True)
plt.show()
```

```python
# Draw box plot to visualize the distribution of the 'attack' data to gain insights
plt.figure(figsize=(10, 6))
plt.boxplot(df2_new['attack'])
plt.xlabel('Attack')
plt.title('Box Plot of Attack Attribute')
plt.show()
```

```python
df1_new = df1.head(10)
```

```python
# Pivot the dataframe for better visualization
pivot_df1 = df1_new.pivot_table(index='NOC', columns='Year', values='Total')
pivot_df1
```

```python
# Plot the heatmap based on pivot table
plt.figure(figsize=(10,6))
sns.heatmap(pivot_df1, annot=True, fmt='d', cmap='crest', linewidth=0.5)
plt.title('Total Medals by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```
