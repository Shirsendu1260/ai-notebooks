```python
from graphviz import Digraph
```

```python
# Create a new directed graph
graph = Digraph()
```

```python
# Define the nodes and edges of the mind map
nodes = {
    'root': ['a','b','c'],
    'a': ['a1','a2'],
    'b': ['b1','b2'],
    'c': ['c1','c2']
}
```

```python
nodes.items()
```

```python
# Add nodes and edges to the graph
for parent, children in nodes.items():
  graph.node(parent)
  for each_child in children:
    graph.node(each_child)
    graph.edge(parent, each_child)
```

```python
# Render and save the mind map
graph.render('mind_map', format='png', cleanup=True)
```

```python
# View the Mind Map
from IPython.display import Image
Image(filename='mind_map.png')
```

Another Example

```python
graph = Digraph()

nodes = {
    'root': ['a','b','c','d'],
    'a': ['a1','a2'],
    'b': ['b1','b2'],
    'c': ['c1','c2'],
    'd':['d1','d2']
}

for parent, children in nodes.items():
  graph.node(parent)
  for each_child in children:
    graph.node(each_child)
    graph.edge(parent, each_child)

graph.render('mind_map', format='png', cleanup=True)

from IPython.display import Image
Image(filename='mind_map.png')
```
