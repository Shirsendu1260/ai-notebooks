```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/synthetic_salary_data.csv')
```

```python
df
```

```python
df.head()
```

```python
# Split into features (x) and target variable (y)
x = df.drop('Salary', axis=1)
y = df['Salary']
```

```python
x
```

```python
y
```

Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# We are calculating MSE and R2 just for our learning purpose. These are applied automatically when the model is applied to minimize error.
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error: ', mse)
print('R-Squared: ', r2)
```

```python
plt.scatter(x_test, y_test, color='black', label='Actual')
plt.plot(x_test, y_pred, color='blue', linewidth=2, label='Predicted')
plt.xlabel('Years of exp.')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Predict the salary for a specific number of years of exp.
year_exp = 5
pred_salary = model.predict(np.array([[year_exp]]))
print(f'Predicted salary for {year_exp} years of experience: ${pred_salary[0]:,.3f}')
```
